```python
# Working perfectly.
def StringChallenge(str):
    seperated = str.split()
    stack = []
    i = 0

    query_first = seperated[0]
    
  # Firsly fill up the stack
    while i <  len(query_first)-1:
#         print(query_first[i])
        if(query_first[i] == '*'): # IF a,b,c... (same thing for 3 time or else {N} tim=numb)
            if query_first[i+1] == "{" :
                numb = int(query_first[i+2])
                for j in range(0,numb):
                    stack.append('*')
                i+=numb
            else:
                stack.append('*')
                stack.append('*')
                stack.append('*')
        elif (query_first[i] == '+'): # IF a,b,c... (occurs once)
            if(query_first[i+1] == '{'):
                numb = int(query_first[i+2])
                for j in range(0,numb):
                    stack.append('+')
                i+=3
            else:
                stack.append('+')
        elif (query_first[i] == '$'): # IF 1,2,3... (occurs once)
            if(query_first[i+1] == '{'):
                numb = int(query_first[i+2])
                for j in range(0,numb):
                    stack.append('$')
                i+=3
            else:
                stack.append('$')
        i+=1
        
    # Now pop the stack
    query_second = seperated[1]
    print("\nstack = ",(stack))
    print("\nquery_second = ",(query_second))
    
    reset = False
    
    print("\nlen(stack) = ", len(stack))
    print("\nlen(query_second = ", len(query_second))
    
    if(len(stack) != len(query_second)):
        print("\nTwo strings does not have same length!\n")
        return False
    
    for i in range(0, len(query_second)):
    # IF it is alphabet
        if (query_second[i].isalpha() and stack[i] == '+'): 
            stack[i] = -1
        elif(query_second[i].isalnum() and stack[i]=='$'): # IF number 
            stack[i] = -1
        elif(stack[i]== '*'):
            j = i
            for j in range(i, len(query_second)-1):   
                if query_second[j+1] != None and stack[j+1] != None:
                    print(query_second[j])
                    if (query_second[j] == query_second[j+1] and stack[j] == '*'): # Still sequence is going om
                        stack[j] = -1
                        if (j+1 == len(query_second)-1):
                            stack[j+1] = -1
                    elif(query_second[j] != query_second[j+1] and stack[j]=='*'): 
                        # Matlab yahan sequence khatam ho raha hai, aur break ker k j+1 se bahir wali for loop
                        # pe dobara se check ho.
                        stack[j] = -1
                        if (j+1 == len(query_second)-1):
                            stack[j+1] = -1
                        break
                i+=j
                
    length = 0
    for i in range(0, len(stack)):
        if stack[i] != -1:
            length+=1
    print("\nlength of stack = ",length)
    print("\nAfter stack = ",(stack))
    if length == 0:
        return True
    else: return False
```
